```python
df = pd.read_csv("/Users/kwaksanghyeok/desktop/소상공인시장진흥공단_상가(상권)정보_서울_202206.csv")
df.head(1)
```

```python
starbucks = df[df.상호명.str.contains('스타벅스')]
starbucks.head(1)
```

```python
ediya = df[df.상호명.str.contains("이디야|EDIYA")]
ediya.head(1)
```

```python
starbucks = starbucks.rename(columns = {'상호명' : '브랜드명'})
starbucks = starbucks.reset_index(drop=True)

starbucks['브랜드명'] = '스타벅스'
```

```python
starbucks = starbucks[['브랜드명', '위도', '경도']]
starbucks
```

```python
ediya = ediya.rename(columns = {'상호명' : '브랜드명'})
ediya = ediya.reset_index(drop=True)

ediya['브랜드명'] = '이디야'
```

```python
ediya = ediya[['브랜드명', '위도', '경도']]
ediya
```

```python
starbucks_ediya = pd.concat([ediya, starbucks], axis = 0) #1은 옆으로 0은 아래로

starbucks_ediya 
```

```python
near_starbucks = []
shortest_distance = []

#과연 for 중첩문을 써야만 했을까?

for i in range(starbucks.shape[0]):
    a = [starbucks["위도"][i], starbucks["경도"][i]]
    for j in range(ediya.shape[0]):
        b = [ediya["위도"][j], ediya["경도"][j]]
        result = haversine(a, b, unit='m')
        if result <= 100 :
            near_starbucks.append(b)
            shortest_distance.append(result)
            print(a,b, result)
```

```python
near_starbucks_ediya = pd.DataFrame(near_starbucks, columns=('위도', '경도'))
```

```python
near_distance = pd.DataFrame(shortest_distance, columns=['최단거리'])
```

```python
pd.options.display.float_format = '{:.13f}'.format
```

```python
near_starbucks_ediya_distance = pd.concat([near_starbucks_ediya, near_distance], axis = 1)
near_starbucks_ediya_distance
```
